# O que é Pandas?

A bilioteca pandas é uma biblioteca de software criada para a linguagem Python para manipulação e análise de dados. Em particular, oferece estruturas de dados e operações para manipular tabelas numéricas e séries temporais. É o principal pacote usado

```python
import pandas as pd
```

```python
dados = pd.read_csv('DataSets/fifa.csv/fifa.csv')
```

```python
dados.loc[dados['Nationality'] == 'Brazil']
```

```python
dados.loc[dados['Name'] == 'Neymar Jr']
```

```python
dados.loc[dados['Age'] == 42]
```

```python
dados.sort_values('Name')
```

```python
dados.sort_values(['Name', 'Age'])
```

```python
dados['Total'] = dados['Acceleration'] + dados['Agility'] + dados['Reactions']
dados
```

```python
dados = dados[['Name', 'Total']]
dados.sort_values('Total', ascending = False)
```

```python
dados.to_csv('dados1.csv', index = False)
```

```python
dados.to_excel('dados2.xlsx', index = False)
```

```python
dados.to_csv('dados3.txt', index = False, sep='\t')
```
